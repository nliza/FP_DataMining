```python
# Distribusi rating (Stars)
sns.histplot(data['Stars'], bins=5, kde=False)
plt.title('Distribusi Stars')
plt.xlabel('Stars')
plt.ylabel('Jumlah')
plt.show()

#Distribusi Usage Period
sns.countplot(y='Usage Period', data=data, order=data['Usage Period'].value_counts().index,)
plt.title('Distribusi Usage Period')
plt.xlabel('Jumlah')
plt.ylabel('Usage Period')
plt.show()

# Distribusi Profile Age
sns.countplot(y='Profile Age', data=data, order=data['Profile Age'].value_counts().index)
plt.title('Distribusi Profile Age')
plt.xlabel('Jumlah')
plt.ylabel('Profile Age')
plt.show()

# Jenis Kombinasi Kulit dari Profile Description
skin_type = data['Profile Description'].str.split(',').str[0]
skin_type_counts = skin_type.value_counts()

# Visualisasi Pie Chart
plt.figure(figsize=(8, 8))
skin_type_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('tab10'))
plt.title('Distribusi Jenis Kulit Pengguna')
plt.ylabel('')
plt.show()


# Distribusi Purchase Point 10 terbesar
top_purchase_points = data['Purchase Point'].value_counts().head(10)

sns.barplot(
    x=top_purchase_points.values,
    y=top_purchase_points.index,
)
plt.title('Distribusi 10 Purchase Point Terbanyak', fontsize=14, fontweight='bold')
plt.xlabel('Jumlah', fontsize=12)
plt.ylabel('Purchase Point', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

# Distribusi Recommend
sns.countplot(x='Recommend', data=data)
plt.title('Distribusi Recommend')
plt.xlabel('Recommend')
plt.ylabel('Jumlah')
plt.show()

#Analisis hubungan Stars vs. Status
sns.boxplot(x='Status', y='Recommend', data=data)
plt.title('Distribusi Rekomendasi Berdasarkan Status')
plt.xlabel('Status')
plt.ylabel('Recommend')
plt.show()

#Analisis hubungan Stars vs. Status
sns.boxplot(x='Status', y='Stars', data=data)
plt.title('Distribusi Stars Berdasarkan Status')
plt.xlabel('Status')
plt.ylabel('Stars')
plt.show()

```

```python
import spacy

nlp = spacy.load("xx_ent_wiki_sm")

data = pd.read_csv('/content/drive/MyDrive/Female Daily/review_FemaleDaily.csv')

def tokenize_text_spacy(text):
    doc = nlp(text)
    return [token.text for token in doc]

data['tokens'] = data['Review'].apply(tokenize_text_spacy)

data.to_csv('/content/drive/MyDrive/Female Daily/review_FemaleDaily_tokenized.csv', index=False)
```

```python
from nltk.corpus import stopwords as nltk_stopwords
import pandas as pd
import ast

stopwords_id = set(nltk_stopwords.words('indonesian'))

data = pd.read_csv('/content/drive/MyDrive/Female Daily/review_FemaleDaily_tokenized.csv')

data['tokens'] = data['tokens'].apply(ast.literal_eval)

def remove_stopwords(tokens):
    return [token for token in tokens if token.lower() not in stopwords_id]

data['stopwords'] = data['tokens'].apply(remove_stopwords)

data.to_csv('/content/drive/MyDrive/Female Daily/review_FemaleDaily_stopwords.csv', index=False)
```
